 ## Análise estatística univariada

```python
#Teste à variância
variancia = merged_metadados.groupby('SEX')['AGE'].var(ddof=0)  # ddof=0 para usar a fórmula da população

print(variancia)
```

- As variâncias de idade calculadas em Python e R para os sexos feminino e masculino estão próximas, mas não são iguais (Em R: Sexo feminino: 110.9004; Sexo masculino: 112.0019). As diferenças entre estas podem ser devido a várias razões, incluindo diferenças nos métodos de cálculo, manipulação de dados ou até mesmo de precisão numérica.

```python
from scipy import stats

# Teste à normalidade
masc = merged_metadados.loc[merged_metadados['SEX'] == 'Male', 'AGE']

stat, p_valor = stats.shapiro(masc)
stat = round(stat, 5)
p_valor = round(p_valor, 5)

print("Estatística de teste:", stat)
print("p-value:", p_valor)
```

```python
fem = merged_metadados.loc[merged_metadados['SEX'] == 'Female', 'AGE']

stat, p_valor = stats.shapiro(fem)
stat = round(stat, 5)
p_valor = round(p_valor, 5)

print("Estatística de teste:", stat)
print("p-value:", p_valor)
```

- Os valores obtidos nos testes de normalidade de Shapiro-Wilk são iguais aos que tinham sido obtidos no R, isto é, encontramos evidências estatísticas que indicam que os dados não se distribuem normalmente (p-value < 0.05) tanto para o sexo masculino quanto para o feminino. Deste modo, foi desenhado em QQplot para ajudar nesta análise:

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt

sm.qqplot(masc, line='s')  # 's' para mostrar a linha de referência para uma distribuição normal padrão
plt.title('Masculino')
plt.show()
```

```python
sm.qqplot(fem, line='s')
plt.title('Feminino')
plt.show()
```

- Ao examinar o gráfico QQ Plot, e comparando-o com a linha diagonal, notamos que a maioria dos pontos, para o sexo masculino, segue uma tendência semelhante à da linha, sugerindo uma possível aproximação à normalidade (igual ao que aconteceu com o R).

```python
from scipy.stats import bartlett

#teste de Bartlett para garantir a homogeneidade das variâncias
statistic, p_value = bartlett(masc,fem)
statistic = round(statistic, 5)
p_value = round(p_value, 5)

print("Bartlett's K-squared:", statistic)
print("p_value:", p_value)
```

- Apesar dos valores no R difererirem um pouco neste teste (valores do R:artlett's K-squared = 0.0037766, p-value = 0.951), ambos os testes apresentaram p-values significativamente altos, o que sugere que não há evidência estatística para rejeitar a hipótese nula de que as variâncias das duas amostras são iguais. O que seignifica que as variâncias parecem ser homogéneas entre as amostras de idade considerando o sexo. Uma vez garantida a homogeneidade, realizamos o teste oneway ANOVA: 

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

modelo_anova = ols('AGE ~ SEX', data=merged_metadados).fit()
anova_resultado = sm.stats.anova_lm(modelo_anova, typ=2)
print(anova_resultado)
```

- Apesar dos valores no R difererirem um pouco neste teste (valores do R: F-value = 4.466, p-value = 0.0352), Ambos os resultados indicam que a variável 'SEX' tem um efeito significativo na variável 'AGE'.

```python
from pydeseq2.dds import DeseqDataSet
from pydeseq2.ds import DeseqStats
```

```python
# Remover as primeiras duas colunas (Hugo_Symbol Entrez_Gene_Id)
mRNA_counts = mRNA_seq.iloc[:, 2:]

# Garantir que todos os valores sejam numéricos
mRNA_counts = mRNA_counts.apply(pd.to_numeric, errors='coerce')

# Filtrando genes
selected_genes = (mRNA_counts > 1).mean(axis=1) > 0.5
mRNA_filtrado = mRNA_counts[selected_genes]
```

```python
# Alinhar as colunas de mRNA_filtrado com os índices de Sexo
merged_metadados = merged_metadados.set_index('SAMPLE_ID')  # Supondo que SampleID seja o identificador de amostra
Sexo = merged_metadados["SEX"]
```

```python
# Garantir que Sexo e mRNA_filtrado tenham os mesmos índices
common_samples = mRNA_filtrado.columns.intersection(Sexo.index)
mRNA_filtrado = mRNA_filtrado[common_samples]
Sexo = Sexo[common_samples]
```

```python
# Criar o DataFrame design com a coluna SEX
design = pd.DataFrame({"SEX": Sexo})
design.index = design.index.astype(str)
```

```python
# Arredondar os valores da matriz de contagem para inteiros
mRNA_counts_rounded = mRNA_filtrado.T.round().astype(int)

# Preparar os dados para DESeq2
dds = DeseqDataSet(
    counts=mRNA_counts_rounded,
    metadata=design,
    design_factors=["SEX"]
)
```

```python
dds.deseq2()
```

```python
stat_res = DeseqStats(dds, contrast = ('SEX', 'Male', 'Female'))
```

```python
stat_res.summary()
```

```python
res = stat_res.results_df
res
```

```python
# Contar o número de genes significantemente regulados
down_genes = res[res['padj'] < 0.05][res['log2FoldChange'] < 0]
up_genes = res[res['padj'] < 0.05][res['log2FoldChange'] > 0]
not_sig_genes = res[res['padj'] >= 0.05]


print("Down:", len(down_genes))
print("NotSig:", len(not_sig_genes))
print("Up:", len(up_genes))
```

Os resultados obtidos em python diferem dos resultados que obtivemos em R:

- Existem 318 genes subexpressos no sexo feminino em comparação com o sexo masculino (R: 3 genes);

- Existem 256 genes sobexpressos no sexo feminino em comparação com o sexo masculino (R: 5 genes);

- 15932 genes não mostraram diferenças significativas de expressão entre os sexos (R: 4119 genes).

```python
import numpy as np

colors = np.where(res['padj'] < 0.05, 'r', 'gray')


plt.figure(figsize=(10, 6))
plt.scatter(res['log2FoldChange'], -np.log10(res['padj']), color=colors, alpha=0.5)
plt.xlabel('Log2 Fold Change')
plt.ylabel('-log10(padj)')
plt.title('Volcano Plot')
plt.axhline(y=-np.log10(0.05), color='gray', linestyle='--', linewidth=1)
plt.axvline(x=0, color='gray', linestyle='--', linewidth=1)
plt.show()
```

Os genes significantemente regulados (com um valor p-ajustado menor que 0.05) estão destacados a vermelho, e os restantes a cinza.

### Clustering

```python
from sklearn import preprocessing
%matplotlib inline
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.stats import ttest_ind

# Preparar os dados
nova = mRNA_seq.iloc[:, 2:].astype(float)

# Filtragem conforme o critério
maximos_nova = nova.max(axis=1)
minimos_nova = nova.min(axis=1)
vl = (maximos_nova / minimos_nova) > 2
nova = nova[vl].dropna()
```

```python
# Ajustar os nomes das linhas
expressao = nova.index.astype(str).str.replace(r"\..*", "", regex=True)

#remover os duplicados
temp_series = pd.Series(expressao)
expressao = temp_series.apply(lambda x: x if temp_series.tolist().count(x) == 1 else f"{x}_1")

# Atualizar nomes das linhas
nova.index = expressao
```

```python
# t-test para selecionar os 50 melhores
_, p_values = ttest_ind(nova.values, np.zeros(nova.shape), axis=1)
ranking = np.argsort(p_values)
p_value = ranking[:50]
melhores_50 = nova.iloc[p_value, :]

gene_names = nova.index[p_value]
melhores_50.index = gene_names
melhores_50
```

```python
from scipy.spatial.distance import pdist

# Matriz de distância e clustering hierárquico
distxy = pdist(melhores_50, metric='euclidean')
Z = linkage(distxy, method='complete')
```

```python
plt.figure(figsize=(25, 10))
dendrogram(
    Z,
    labels=melhores_50.index,
    leaf_rotation=90.,  # rotaciona os rótulos no eixo x
    leaf_font_size=8.,  # tamanho da fonte para os rótulos no eixo x
)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample index')
plt.ylabel('Distance')
plt.show()
```

```python
gene_names = nova.index[p_value]
melhores_50.index = gene_names

# Criar o heatmap usando seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(melhores_50, cmap="viridis", yticklabels=True)

# Adicionar títulos e rótulos
plt.title('Heatmap dos 50 Melhores Genes')
plt.xlabel('Amostras')
plt.ylabel('Genes')

# Mostrar o heatmap
plt.show()
```

### K-means

```python
from sklearn.cluster import KMeans

nova = nova.iloc[:len(merged_metadados['SEX'])]

#assegurar reprodutibilidade
np.random.seed(123)

# Realizar o k-means clustering para diferentes números de clusters (2 a 10)
ofs = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=123)
    kmeans.fit(nova)
    ofs.append(kmeans.inertia_)
    
plot_data = pd.DataFrame({'num_clusters': range(2, 11), 'wss': ofs})
plt.figure(figsize=(10, 6))
sns.lineplot(x='num_clusters', y='wss', data=plot_data, marker='o')
plt.xlabel('Número de Clusters')
plt.ylabel('WSS')
plt.title('WSS vs. Número de Clusters')
plt.show()
```

À semelhança do que aconteceu no R:
- A partir de k=4, a redução da WSS diminui significativamente, indicando que adicionar mais clusters após esse ponto não traz uma melhoria substancial na compactação dos clusters.

```python
from sklearn.preprocessing import KBinsDiscretizer

# Realizar o k-means clustering com 4 clusters
kmeans = KMeans(n_clusters=4, n_init=10, random_state=123)
clusters = kmeans.fit_predict(nova)

# Converter os clusters em uma categoria com rótulos
centers_factor = pd.Categorical(clusters, categories=[0, 1, 2, 3], ordered=True)
centers_factor = centers_factor.rename_categories(['Cluster 1', 'Cluster 2', 'Cluster 3', 'Cluster 4'])

# Obter o fator sexo dos metadados
sex_factor = merged_metadados['SEX'].astype('category')

# Criar a tabela de contingência
table = pd.crosstab(centers_factor, sex_factor)

# Exibir a tabela de contingência
print(table)
```

- Cluster 1: é o que possui maior número de mulheres (106) e homens (298), predominantemente, é composto por homens, com uma proporção significativamente maior em relação às mulheres;
- Cluster 2: temos apenas 1 homem;
- Cluster 3: apenas 1 mulher;
- Cluster 4: apenas 1 homem.


Com base nestes resultados, pode-se concluir que os clusters formados pelo algoritmo K-means têm diferentes distribuições de sexo.

### k-Nearest Neighbor

```python
data_train = data_train.copy()
data_teste = data_teste.copy()
np.random.seed(123456)

# Inicialização do LabelEncoder
le = LabelEncoder()

# Ajustar e transformar a coluna 'TUMOR_STATUS' nos dados de treino
data_train['TUMOR_STATUS'] = le.fit_transform(data_train['TUMOR_STATUS'])

# Verificar se todas as classes estão presentes no conjunto de teste
missing_classes = set(le.classes_) - set(data_teste['TUMOR_STATUS'])
for cls in missing_classes:
    data_teste = data_teste.append(pd.Series([np.nan] * len(data_teste.columns), index=data_teste.columns), ignore_index=True)
    data_teste.at[data_teste.index[-1], 'TUMOR_STATUS'] = cls

data_teste['TUMOR_STATUS'] = le.transform(data_teste['TUMOR_STATUS'])

# Remover valores ausentes
data_train = data_train.dropna()
data_teste = data_teste.dropna()

# Separar características
X_train = data_train.iloc[:, :-1].values
y_train = data_train.iloc[:, -1].values
X_test = data_teste.iloc[:, :-1].values
y_test = data_teste.iloc[:, -1].values

# Certificar-se de que os arrays são C-contiguous
X_train = np.ascontiguousarray(X_train)
X_test = np.ascontiguousarray(X_test)


knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
knn_pred = knn.predict(X_test)


conf_matrix = confusion_matrix(y_test, knn_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=le.classes_)
disp.plot()
```

 A matriz criada tem quadrantes diferentes:

- Verdadeiro negativo (quadrante superior esquerdo)
- Falso positivo (quadrante superior direito)
- Falso negativo (quadrante inferior esquerdo)
- Verdadeiro positivo (quadrante inferior direito)

 Assim sendo, retiramos as seguintes conclusões:
- O modelo classificou corretamente 59 casos como ‘TUMOR FREE’ (VN);
- O modelo classificou incorretamente 17 casos como ‘TUMOR FREE’, quando na verdade eram ‘WITH TUMOR’(FP);
- O modelo classificou incorretamente 32 casos reais de ‘WITH TUMOR’ como ‘TUMOR FREE’ (FN);
- Identificou 6 casos reais de ‘WITH TUMOR’ (VP);

```python
from sklearn import metrics

np.random.seed(123456)
Accuracy = metrics.accuracy_score(y_test, knn_pred)
Precision = metrics.precision_score(y_test, knn_pred, average='macro')
Sensitivity_recall = metrics.recall_score(y_test, knn_pred, average='macro')
Specificity = metrics.recall_score(y_test, knn_pred, average='macro', pos_label=0)

print(f"Accuracy: {Accuracy:.2f}")
print(f"Precision: {Precision:.2f}")
print(f"Sensitivity_recall: {Sensitivity_recall:.2f}")
print(f"Specificity: {Specificity:.2f}")
```
